```python
travels_dataset['Purpose Of Travel'] = travels_dataset['Purpose Of Travel'].fillna('Not Provided')
```

```python
travels_dataset.info()
```

```python
def extract_city(text):
    result = text.split(',')[0]
    return result
    
```

```python
def extract_state(text):
    result = text.split(',')[1]
    return result
    
```

```python
travels_dataset['State'] = travels_dataset['Destination'].apply(extract_state)
```

```python
travels_dataset['City'] = travels_dataset['Destination'].apply(extract_city)
```

```python
def extract_firstname(text):
    result = text.split(' ')[0]
    return result
```

```python
def extract_lastname(text):
    result = text.split(' ')[-1]
    return result
```

```python
travels_dataset['Firstname'] = travels_dataset['Employee'].apply(extract_firstname)
```

```python
travels_dataset['Lastname'] = travels_dataset['Employee'].apply(extract_lastname)
```

```python
travels_dataset
```

```python
travels_dataset.drop(columns='Lasttname')
```

```python
columns = ['Employee', 'Travel Start Date', 'Travel End Date', 'Actual Total Expenses', 'Purpose Of Travel']

final_result = travels_dataset[columns].copy()
```

```python
final_result.to_csv('cleaned_travels_dataset.csv')
```

```python
def remove_special_characters(name):
    result = name.replace('@', '')
    result = result.replace('!', '')
    return result
```

```python
travels_dataset['Employee'] = travels_dataset['Employee'].apply(remove_special_characters)
```

```python
travels_dataset.to_csv('cleaned_travels_dataset.csv')
```
